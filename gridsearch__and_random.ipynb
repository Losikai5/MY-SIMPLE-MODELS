```python
X = breast_cancer.data
y = breast_cancer.target
```

```python
print(X.shape)
print(y.shape)
```

```python
print(y)
```

```python
print(X)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python

para = {
    'C':[ 0.1, 1,10,100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}
```

```python
clf =  GridSearchCV(model, para,cv=5)
```

```python
clf.fit(X,y)
```

```python
clf.cv_results_
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.cv_results_
```

```python
pa2 = {
    'C':[ 0.1, 1,10,100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

```

```python
clf2 = RandomizedSearchCV(model, pa2, cv=5)
```

```python
clf2.fit(X,y)
```

```python
clf2.best_params_
```

```python
clf2.best_score_
```

```python
clf2.cv_results_
```
