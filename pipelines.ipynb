```python
from sklearn.preprocessing import StandardScaler
data.head()
```

```python
num_cols = ['R&D Spend', 'Administration', 'Marketing Spend']
cat_cols = ['State']
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Profit'], axis = 1)
y = data['Profit']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
num = Pipeline(steps=[('scaler', StandardScaler())])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
cat = Pipeline(steps=[('encoder', OneHotEncoder(handle_unknown = 'ignore',sparse_output=False))])
```

```python
trans = ColumnTransformer(transformers=[('num', num, num_cols), ('cat', cat, cat_cols)],remainder='drop',n_jobs=-1)
```

```python
final = Pipeline(steps=[('preprocessor', trans), ('regressor', LinearRegression())])
```

```python
final.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
final.score(X_test, y_test)
```

```python
r2_score(y_test, final.predict(X_test))
```
