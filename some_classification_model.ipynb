```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head()

```

```python
df['Pregnancies'] = df['Pregnancies'].replace(0,np.nan)
```

```python
df['Glucose'] = df['Glucose'].replace(0,np.nan)
```

```python
df['BloodPressure'] = df['BloodPressure'].replace(0,np.nan)
```

```python
df['SkinThickness'] = df['SkinThickness'].replace(0,np.nan)
```

```python
df['Insulin'] = df['Insulin'].replace(0,np.nan)
```

```python
df['BMI'] = df['BMI'].replace(0,np.nan)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.head(40)
```

```python
df.describe()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from xgboost import XGBClassifier
```

```python
clf = XGBClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc = accuracy_score(y_test,y_pred)
print(acc*100)
```

```python
con = confusion_matrix(y_test,y_pred)
print(con)
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
kflod = KFold(n_splits=10,random_state=7,shuffle=True)
```

```python
result = cross_val_score(clf,X,y,cv=kflod)
print(result.mean()*100)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'learning_rate': [0,0.01, 0.1, 0.2,1],
    'max_depth': [3, 4, 5],
    'n_estimators': [50, 100,40,60],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'subsample': [0.8, 0.9, 1.0]
}
```

```python
search = GridSearchCV(clf,param_grid,scoring='accuracy',n_jobs=-1,cv=3)
```

```python
search.fit(X_train,y_train)
print("best parameters =",search.best_params_)
print("best score = ",search.best_score_*100)
```

```python

best_model = search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
